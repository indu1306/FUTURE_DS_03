```python
print("Code started")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\RAMYA SRI\Downloads\marketing_funnel_project\2019-Nov.csv"
df = pd.read_csv(file_path, nrows=100000)


print("\nFirst 5 rows:")
print(df.head())

print("\nShape of data:", df.shape)
print("\nColumns in dataset:", df.columns)


df.dropna(subset=['event_type'], inplace=True)


visitors = df[df['event_type'] == 'view']['user_id'].nunique()
leads = df[df['event_type'] == 'cart']['user_id'].nunique()
customers = df[df['event_type'] == 'purchase']['user_id'].nunique()

print("\n--- Funnel Results ---")
print("Visitors:", visitors)
print("Leads:", leads)
print("Customers:", customers)


drop1 = visitors - leads
drop2 = leads - customers

print("\n--- Drop-off ---")
print("Drop (Visitors → Leads):", drop1)
print("Drop (Leads → Customers):", drop2)


conversion_v_to_l = (leads / visitors) * 100 if visitors != 0 else 0
conversion_l_to_c = (customers / leads) * 100 if leads != 0 else 0

print("\n--- Conversion Rates ---")
print("Visitors → Leads (%):", round(conversion_v_to_l, 2))
print("Leads → Customers (%):", round(conversion_l_to_c, 2))


funnel = ['Visitors', 'Leads', 'Customers']
values = [visitors, leads, customers]

plt.figure()
plt.bar(funnel, values)
plt.title("Marketing Funnel Analysis")
plt.xlabel("Stages")
plt.ylabel("Number of Users")
plt.show()


drop_values = [drop1, drop2]
stages = ['Visitors → Leads', 'Leads → Customers']

plt.figure()
plt.bar(stages, drop_values)
plt.title("Drop-off Analysis")
plt.xlabel("Stages")
plt.ylabel("Users Lost")
plt.show()

print("\nCode finished successfully ")
```
